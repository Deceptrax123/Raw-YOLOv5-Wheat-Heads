```python
#notebook of all code experiments
import pandas as pd 
import numpy as np 
```

```python
pred=pd.read_csv("YOLOv5_config/utils/predictions.csv")
sample=pd.read_csv("/Volumes/T7 Shield/Smudge/Datasets/Wheat_Head_detection/submission.csv")
```

```python
pred.shape
```

```python
pred_names=list(pred['image_name'])
pred_boxes=list(pred['BoxString'])
sample_names=list(sample['image_name'])
```

```python
len(sample_names)
len(pred_names)
```

```python
for i in sample_names:
    if i not in pred_names:
        pred_names.append(i)
        pred_boxes.append('no_box')
```

```python
pred_new=pd.DataFrame({
    'image_name':pred_names,
    'BoxString':pred_boxes
})

pred_new.to_csv("pred_edited.csv",)
```
